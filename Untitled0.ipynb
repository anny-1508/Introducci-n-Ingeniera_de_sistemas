```python
# Ejercicio No 17
import math

radio = float(input("Ingrese el radio del círculo: "))

area = math.pi * radio ** 2
longitud = 2 * math.pi * radio

print("El área del círculo es:", round(area, 2))
print("La longitud de la circunferencia es:", round(longitud, 2))
```
